```python
import urllib.request
from bs4 import BeautifulSoup
from collections import Counter
import re


url = "https://www.w3schools.com/"
def fetch_and_parse(url):
    try:
        response = urllib.request.urlopen(url)
        html_content = response.read()

        
        soup = BeautifulSoup(html_content, "html.parser")

    
        text = soup.get_text()

        return text
    except Exception as e:
        print("An error occurred:", str(e))
        return None


def calculate_word_frequencies(text):
    if text is None:
        return
    words = re.findall(r'\w+', text.lower())
    word_freq = Counter(words)
    for word, freq in word_freq.most_common():
        print(f"{word}: {freq}")

if __name__ == "__main__":
    
    web_page_text = fetch_and_parse(url)

    
    calculate_word_frequencies(web_page_text)
```

```python
import urllib.request
from bs4 import BeautifulSoup


url = "https://www.w3schools.com/"


def fetch_and_parse(url):
    try:
    
        response = urllib.request.urlopen(url)
        html_content = response.read()

    
        soup = BeautifulSoup(html_content, "html.parser")

        return soup
    except Exception as e:
        print("An error occurred:", str(e))
        return None


def display_hyperlinks(soup):
    if soup is None:
        return

    
    anchor_tags = soup.find_all("a")

    
    for tag in anchor_tags:
        href = tag.get("href")
        if href:
            print(href)

if __name__ == "__main__":

    parsed_html = fetch_and_parse(url)

    
    display_hyperlinks(parsed_html)
```

```python
import json


student_marks = {
    "Alice": 92,
    "Bob": 78,
    "Charlie": 85,
    "David": 95,
    "Eve": 88
}

num_students = len(student_marks)


with open("student_marks.json", "w") as json_file:
    json.dump(student_marks, json_file)

with open("student_marks.json", "r") as json_file:
    data = json.load(json_file)
    print(f"Number of students: {num_students}")
    for student, marks in data.items():
        print(f"Student: {student}, Marks: {marks}")
```

```python
import requests
from bs4 import BeautifulSoup

url = "https://weather.com/weather/7day/l/Tiruchirapalli+TN+INXX0292:1:IN"


response = requests.get(url)

if response.status_code == 200:
    soup = BeautifulSoup(response.content, 'html.parser')
    
    forecast_container = soup.find("div", class_="DailyForecast--DisclosureList--u0yMq")
    
    for day in forecast_container.find_all("div", class_="Disclosure--Disclosure--Cwa2d"):
        date = day.find("span", class_="DetailsSummary--daypartDate--1Mebr").get_text(strip=True)
        description = day.find("div", class_="DetailsSummary--dayDescription--1sxN1").get_text(strip=True)
        high_temp = day.find("span", class_="DetailsSummary--highTempValue--3x6cL").get_text(strip=True)
        low_temp = day.find("span", class_="DetailsSummary--lowTempValue--1DlJK").get_text(strip=True)
        
        print(f"Date: {date}")
        print(f"Description: {description}")
        print(f"High Temperature: {high_temp}")
        print(f"Low Temperature: {low_temp}")
        print()
else:
    print("Failed to retrieve the weather data.")

```

```python
import yfinance as yf
import matplotlib.pyplot as plt


stock_symbols = ["INFY", "CTS"]
start_date = "2022-09-19"  
end_date = "2023-09-19"    


stock_data = {}


for symbol in stock_symbols:
    stock_data[symbol] = yf.download(symbol, start=start_date, end=end_date)
plt.figure(figsize=(12, 6))

for symbol, data in stock_data.items():
    plt.plot(data.index, data["Adj Close"], label=symbol
plt.title("Stock Price Movements")
plt.xlabel("Date")
plt.ylabel("Price (USD)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
